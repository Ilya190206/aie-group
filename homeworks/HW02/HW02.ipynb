```python
print("Доля пропусков в каждом столбце:")
missing_percent = df.isna().mean() * 100
print(missing_percent[missing_percent > 0].round(2).astype(str) + '%')
if missing_percent.sum() == 0:
    print("Пропусков нет")


print("\nПроверка дубликатов:")
duplicates = df.duplicated().sum()
print(f"Полностью дублирующих строк: {duplicates}")
if duplicates > 0:
    print("Дубликаты:")
    print(df[df.duplicated()])


print("\nПоиск подозрительных значений:")

numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if (df[col] < 0).any():
        negative_count = (df[col] < 0).sum()
        print(f"В столбце '{col}' найдено {negative_count} отрицательных значений")


print("\nУникальные значения в каждом столбце (первые 10):")
for col in df.columns:
    unique_vals = df[col].dropna().unique()
    if len(unique_vals) <= 10:
        print(f"{col}: {sorted(unique_vals)}")
    else:
        print(f"{col}: {len(unique_vals)} уникальных значений")
        

print("\nСтатистика числовых столбцов (мин/макс/среднее):")
for col in numerical_cols:
    print(f"{col}: min={df[col].min():.2f}, max={df[col].max():.2f}, mean={df[col].mean():.2f}")
```

```python
Обнаруженные проблемы качества данных:
При анализе датасета были выявлены следующие проблемы во-первых, обнаружены пропущенные значения в нескольких столбцах, что может повлиять на результаты анализа и требовать стратегии обработки (удаление или заполнение) Во-вторых, найдено определенное количество полностью дублирующих строк, которые необходимо устранить для избежания искажения статистик
Кроме того, в числовых полях присутствуют подозрительные значения: отрицательные величины там, где они не имеют смысла (например, в показателях количества или цены), а также нереалистичные данные (экстремально высокие значения в возрастных полях или нулевые показатели там, где они должны быть положительным Эти аномалии требуют дополнительной проверки и возможной коррекции перед дальнейшим анализом.
```

```python
import pandas as pd
import numpy as np

filepath = "https://raw.githubusercontent.com/Ilya190206/aie-group/refs/heads/main/homeworks/HW02/S02-hw-dataset.csv"
df = pd.read_csv(filepath)
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    for col in categorical_cols[:2]:
        print(f"\nСтолбец: {col}")
        freq = df[col].value_counts()
        print("Частоты:")
        print(freq.head(10))
        print(f"\nУникальных значений: {df[col].nunique()}")
        print(f"Самое частое: {freq.index[0]} ({freq.iloc[0]} раз)")
else:
    print("Категориальных столбцов не найдено")
    
numeric_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(include=['object']).columns

if len(cat_cols) > 0 and len(numeric_cols) > 0:
   
    group_col = cat_cols[0]
    agg_cols = list(numeric_cols[:2])
    
    print(f"Группировка по столбцу: '{group_col}'")
    print(f"Агрегация по столбцам: {agg_cols}")
    
    grouped = df.groupby(group_col)[agg_cols].agg(['mean', 'sum', 'count'])
    print("\nРезультат группировки:")
    print(grouped.round(2))
    
  
    if 'age' in df.columns and 'income' in df.columns:
        print("\nДополнительная группировка (если есть возраст и доход):")
        age_income_stats = df.groupby('age')['income'].agg(['mean', 'median', 'count']).round(2)
        print(age_income_stats.head(10))
else:
    print("Недостаточно столбцов для группировки")



if len(numeric_cols) > 0:
 
    num_col = numeric_cols[0]
    print(f"Создание коридоров для столбца: '{num_col}'")
    
    if len(df[num_col].dropna()) > 0:
        
        df[f'{num_col}_bins'] = pd.cut(df[num_col], bins=5)
        
        print("Распределение по коридорам:")
        bins_dist = df[f'{num_col}_bins'].value_counts().sort_index()
        print(bins_dist)
        
        
        if len(numeric_cols) > 1:
            second_num_col = numeric_cols[1]
            bins_analysis = df.groupby(f'{num_col}_bins')[second_num_col].agg(['mean', 'count']).round(2)
            print(f"\nАнализ '{second_num_col}' по коридорам '{num_col}':")
            print(bins_analysis)
    else:
        print(f"Столбец '{num_col}' пустой или содержит только NaN")
else:
    print("Числовых столбцов не найдено")

if len(numeric_cols) >= 2:
    
    corr_matrix = df[numeric_cols[:min(5, len(numeric_cols))]].corr()
    print(corr_matrix.round(3))
```

```python
Наблюдения:

1. Категории доминируют: Больше всего пользователей из России (RU, 13 человек, 29%) и Франции (FR, 12 человек, 27%). Вместе они составляют 56% всех данных.

2. Отличия групп по средним: Пользователи из Франции в среднем старше всех - 46 лет, а из Китая моложе всех - 24 года. Немецкие пользователи имеют самые высокие номера user_id (в среднем 24.7), что может указывать на более позднюю регистрацию, а российские - самые низкие (18.1).

3. Неожиданные эффекты: Обнаружена обратная зависимость - чем старше пользователь, тем меньше покупок он совершает (корреляция -0.35). Нет связи между возрастом и временем регистрации. Интересно, что пользователи с самыми первыми номерами user_id (от 0.96 до 8.8) имеют самый высокий средний возраст - 47 лет.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

save_path = r'C:\Users\ulya-\Documents\aie-group\homeworks\HW02\figures'
os.makedirs(save_path, exist_ok=True)

numeric_cols = df.select_dtypes(include=[np.number]).columns


if len(numeric_cols) > 0:
    col = numeric_cols[0]
    
    plt.figure(figsize=(10, 6))
    plt.hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
    plt.xlabel(col)
    plt.ylabel('Частота')
    plt.title(f'Распределение {col}')
    plt.grid(True, alpha=0.3)
    plt.savefig(os.path.join(save_path, 'histogram.png'), dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Гистограмма сохранена: {os.path.join(save_path, 'histogram.png')}")


if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 6))
    cols_to_plot = numeric_cols[:min(5, len(numeric_cols))]
    data_to_plot = [df[col].dropna() for col in cols_to_plot]
    plt.boxplot(data_to_plot, labels=cols_to_plot)
    plt.ylabel('Значения')
    plt.title('Боксплот числовых признаков')
    plt.grid(True, alpha=0.3)
    plt.savefig(os.path.join(save_path, 'boxplot.png'), dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Боксплот сохранен: {os.path.join(save_path, 'boxplot.png')}")


if len(numeric_cols) >= 2:
    x_col, y_col = numeric_cols[0], numeric_cols[1]
    
    plt.figure(figsize=(10, 6))
    plt.scatter(df[x_col], df[y_col], alpha=0.6, s=50)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'{x_col} vs {y_col}')
    plt.grid(True, alpha=0.3)
    plt.savefig(os.path.join(save_path, 'scatter_plot.png'), dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Scatter plot сохранен: {os.path.join(save_path, 'scatter_plot.png')}")

print("\nВсе графики сохранены в папку:", save_path)
```
