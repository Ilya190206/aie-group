```python
2. Примеры создания массивов:
```

```python
import numpy as np
list_array=np.array([1,2,3,4,5]) #создание массива из списка
print(list_array)
```

```python
import numpy as np
zeros_array=np.zeros(5) # создание массива с помощью zeros
print(zeros_array)
```

```python
import numpy as np
ones_array=np.ones(5) # создание массива с помощтю zeros
print(ones_array)
```

```python
import numpy as np
arange_array=np.arange(5) # создание массива с помощью arange
print(arange_array)
```

```python
3. Демонстрация атрибутов:
```

```python
import numpy as np
demo_array=np.array([1,2,3,4],dtype=int) # создание массива и указ типа данных
print(demo_array)
print(demo_array.shape) # форма 
print(demo_array.dtype) # тип данных 
```

```python
4. Примеры индексирования и срезов:
```

```python
import numpy as np
array=np.array([0,1,2,3,4,5])
print(array[1]) # вывод элемента массива с индексом 1
print(array[1:3]) # вывод подмассива с помощью среза
```

```python
5. Примеры векторных операций и базовых статистик:
```

```python
import numpy as np
array1=np.array([1,3,8,10])
array2=np.array([1,2,8,6])
print(array1+array2) # поэлементное сложение массивов
print(array1*array2) # поэлементное умножение массивов
print(np.mean(array1)) # вычисление средней величины 1 массива
print(np.mean(array2))# вычесление средней веоичины 2 массива
```
